## Importing the libraries

```python
#Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
os.chdir(os.path.dirname(os.getcwd()))
```

## Segregating the Graduate Courses from Undergrad Courses

```python
#Loading the initial data sets
df_fall = pd.read_excel('data/df_fall_all.xlsx')
df_spring = pd.read_excel('data/df_spring_all.xlsx')
```

```python
df_fall.head()
```

```python
df_spring.head()
```

```python
# Segregating the grad from undergrad for Fall semester

df_fall['Catalog Number'] = pd.to_numeric(df_fall['Catalog Number'], errors='coerce')

# Filter the data where Catalog Number is greater than or equal to 500
df_grad_fall = df_fall[df_fall['Catalog Number'] >= 500]

# Save the filtered grad fall dataset to a new Excel file
df_grad_fall_file_path = 'data/df_grad_fall.xlsx'
df_grad_fall.to_excel(df_grad_fall_file_path, index=False)

# Display the filtered data
df_grad_fall.head()
```

```python
# Segregating the grad from undergrad for Spring semester

df_spring['Catalog Number'] = pd.to_numeric(df_spring['Catalog Number'], errors='coerce')

# Filter the data where Catalog Number is greater than or equal to 500
df_grad_spring = df_spring[df_spring['Catalog Number'] >= 500]

# Save the filtered grad spring dataset to a new Excel file
df_grad_spring_file_path = 'data/df_grad_spring.xlsx'
df_grad_spring.to_excel(df_grad_spring_file_path, index=False)

# Display the filtered data
df_grad_spring.head()
```

```python
# Combine the Fall and Spring grad datasets
df_grad_combined = pd.concat([df_grad_fall, df_grad_spring], ignore_index=True)

# Save the combined dataset to a new Excel file
df_grad_combined_file_path = 'data/df_grad_combined.xlsx'
df_grad_combined.to_excel(df_grad_combined_file_path, index=False)

# Display the file path for download
df_grad_combined_file_path
df_grad_combined
```

## Check for missing values

```python
# Checking for missing values in each column
print("Missing values in each column:")
print(df_grad.isnull().sum())
```

## Check for Duplicates

```python
# Number of duplicate rows
num_duplicates = df_grad.duplicated().sum()
print("Number of duplicate rows:", num_duplicates)
```

```python
# Removeing duplicate rows and saving this as df
if num_duplicates > 0:
    df = df_grad.drop_duplicates()
    print("Duplicates dropped. New shape:", df.shape)
```

## Multivariate Analysis

```python
# Pair Plot of All Numeric Variables
# This plot gives an overall view of relationships among all numeric features.

# Define the list of numerical columns
numerical_columns = df.select_dtypes(include=[np.number]).columns.tolist()

# Create the pair plot
sns.pairplot(df[numerical_columns])
plt.suptitle("Pair Plot of Numeric Variables", y=1.02)
plt.show()

```

## Categorical Analysis

```python
# Frequency and Bar Charts for Categorical Variables
# Looping through all categorical variables to display their frequency counts
categorical_columns = df.select_dtypes(include=['object']).columns.tolist()

for col in categorical_columns:
    # Print frequency counts
    freq = pd.DataFrame(df[col].value_counts())
    print(f"Frequency counts for '{col}':")
    print(freq)
    print("\n")


       # Plot bar chart if there are not too many categories
    if len(freq) < 15:
        plt.figure(figsize=(7, 4))
        freq.plot(kind='bar', color='skyblue', edgecolor='black')
        plt.title(f"Bar Chart of {col}")
        plt.xlabel(col)
        plt.ylabel("Count")
        plt.xticks(rotation=45)
        plt.grid(axis='y')
        plt.show()
```

## Findings

## Overall Findings

The dataset originally contained **1,305 graduate-level course offerings,** out of which **58 duplicate rows** were identified and removed. After cleaning, the final dataset consists of **1,247 unique records,** each representing a distinct course offering across various semesters and instructional modes.

**Column Types:**

**Numerical:** 8 columns (e.g., Total Enroll, Enrollment Capacity)

**Categorical:** 18 columns (e.g., Term, Campus, Component)

**Datetime:** 2 columns (Start Date, End Date)

## Findings by Column
Here are some insights from specific columns:

### Numerical Columns:

**Total Enroll:**

Mean: ~6 students

Median: 2 students

Max: 102 students — highly skewed with many low-enrollment courses

**Enrollment Capacity:**

Mean: ~22, Median: 25 — most classes have moderate capacity

**Room Capacity:**

Median: 1 — many online or small-format courses

Max: 145 — indicates a few large in-person sessions

**Catalog Number:**

Range: 501 to 920 — focused on graduate-level courses (≥ 500)

### Categorical Columns:

**Term:** Spread across 8 terms (Fall 2021 to Spring 2025), with Spring 2025 having the highest count (191)

**Campus:**

69% are from Main Campus

31% are from Arizona Online

**Component:**

Predominantly Lecture (881)

Followed by Independent Study (359)

**Instruction Mode:**

60% are Fully Online

38% In-Person

A small number are Live Online or Hybrid

**Meeting Days & Time:**

A majority of courses have undefined (-) meeting days and times — possibly asynchronous online formats
